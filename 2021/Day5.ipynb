```polyglot-notebook
int CountOverlappingPoints(string[] inputLines)
{
    var allPoints = inputLines.SelectMany(ExpandLine);

    return allPoints
        .GroupBy(p => p)
        .Where(g => g.Count() > 1)
        .Count();

}

IEnumerable<Point> ExpandLine(string inputLine)
{
    var (x1, y1, x2, y2) = inputLine.ParseInts().ToArray();

    if ((x1 != x2) && (y1 != y2)) { yield break; }

    var (xStart, xEnd) = MinMax([x1, x2]);
    var (yStart, yEnd) = MinMax([y1, y2]);

    foreach (var x in Range(xStart, 1 + xEnd - xStart))
    foreach (var y in Range(yStart, 1 + yEnd - yStart))
    {
        yield return (x, y);
    }
}
```

```polyglot-notebook
// In the above example, this is anywhere in the diagram with a 2 or larger - a
// total of 5 points.

var testAnswer = CountOverlappingPoints(testInputLines);
Console.WriteLine(testAnswer);
```

```polyglot-notebook
// Consider only horizontal and vertical lines. At how many points do at least
// two lines overlap?

var part1Answer = CountOverlappingPoints(inputLines);
Console.WriteLine(part1Answer);
```

Ok, we could modify part 1 to include a toggle as to whether to include diagonals, but just for funsies let's leave it alone and create a diagonal-only point enumerator, then we can concat them together.

```polyglot-notebook
IEnumerable<Point> ExpandDiagonal(string inputLine)
{
    var (x1, y1, x2, y2) = inputLine.ParseInts().ToArray();

    if (x1 == x2 || y1 == y2) { return []; }

    var (xStart, xEnd) = MinMax([x1, x2]);
    var distance = xEnd - xStart + 1;

    var xDir = x2 < x1 ? -1 : 1;
    var yDir = y2 < y1 ? -1 : 1;
    Point dir = (xDir, yDir);

    return Range(0, distance).Select(i => (x1, y1) + (dir * i));
}

int CountOverlappingPoints2(string[] inputLines)
{
    var allPoints = inputLines.SelectMany(ExpandLine)
        .Concat(inputLines.SelectMany(ExpandDiagonal));

    return allPoints
        .GroupBy(p => p)
        .Where(g => g.Count() > 1)
        .Count();
}
```

```polyglot-notebook
var part2Answer = CountOverlappingPoints2(inputLines);
Console.WriteLine(part2Answer);
```
