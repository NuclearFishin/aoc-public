Curiously, the sample input and our puzzle input have an even number of input lines, so in theory `0`s and `1`s could occur in equal frequency, but the puzzle description doesn't specifically acknowledge that case! It's conspicuous absence feels deliberate to me, so I'm going to assume this just doesn't happen, indeed it doesn't for the puzzle input.

Another thing that jumps out is that since the gamma rate is the most common digit, and epsilon is the least common, the epsilon rate is just the binary inverse of the gamma.

```polyglot-notebook
int GetPowerConsumption(string[] inputLines)
{
    var count = inputLines.Length;
    var digits = inputLines[0].Length;
    var mid = count / 2;

    var mostStr = new string(Range(0, digits).Select(getMost).ToArray());

    var gammaVal = BaseNToInt(mostStr, "01");
    var epsilonVal = BaseNToInt(mostStr, "10");

    return gammaVal * epsilonVal;

    char getMost(int pos) => inputLines.Count(l => l[pos] is '1') > mid ? '1' : '0';
}
```

```polyglot-notebook
// Multiplying the gamma rate (22) by the epsilon rate (9) produces the power
// consumption, 198.

var testAnswer = GetPowerConsumption(testInputLines);
Console.WriteLine(testAnswer);
```

```polyglot-notebook
// What is the power consumption of the submarine? 

var part1Answer = GetPowerConsumption(inputLines);
Console.WriteLine(part1Answer);
```

Ok, now we get the acknowledgement that we could have an equally common number of `0`s and `1`s :) We'll deal with that accordingly.

For part 2 we still need two intermediate values but they are no longer inverses of each other, so we'll need to generate those intermediate binary strings for the oxygen genrator and CO2 scrubber. From there, the process of converting binary to decimal is straightforward.

```polyglot-notebook
string FindRatingNumber(string[] inputLines, bool keepMost)
{
    var remainLines = inputLines.ToList();
    var digits = remainLines[0].Length;
    var checkPos = 0;

    while (remainLines.Count > 1)
    {
        var count = remainLines.Count;
        var mid = (count + 1) / 2; // bias toward the majority for odd counts

        var ones = remainLines.Count(l => l[checkPos] is '1');
        var mostChar = ones >= mid ? '1' : '0';
        var removeChar = keepMost ? Flip(mostChar) : mostChar;

        var removed = remainLines.RemoveAll(line => line[checkPos] == removeChar);
        if (removed is 0) { throw new Exception("Loop variant failed"); }
        checkPos++;
    }

    return remainLines.Single();
}

int CalcLifeSupportRating(string[] inputLines)
{
    var o2NumberBinary = FindRatingNumber(inputLines, keepMost: true);
    var co2NumberBinary = FindRatingNumber(inputLines, keepMost: false);

    var o2Number = BaseNToInt(o2NumberBinary, "01");
    var co2Number = BaseNToInt(co2NumberBinary, "01");

    return o2Number * co2Number;
}

char Flip(char c) => c is '0' ? '1' : '0';
```

```polyglot-notebook
// Finally, to find the life support rating, multiply the oxygen generator
// rating (23) by the CO2 scrubber rating (10) to get 230.

var part2TestResult = CalcLifeSupportRating(testInputLines);
Console.WriteLine(part2TestResult);
```

```polyglot-notebook
// Use the binary numbers in your diagnostic report to calculate the oxygen
// generator rating and CO2 scrubber rating, then multiply them together. What
// is the life support rating of the submarine? 

var part2Answer = CalcLifeSupportRating(inputLines);
Console.WriteLine(part2Answer);
```
