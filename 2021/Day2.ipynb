```polyglot-notebook
string[] testInputLines = [
    "forward 5",
    "down 5",
    "forward 8",
    "up 3",
    "down 8",
    "forward 2",
];
```

```polyglot-notebook
// Ok, we only have forward, down and up. I think we can sum the related movements and subtract up from down
```

```polyglot-notebook
int DoMovements(string[] inputLines)
{
    int moveAll(string movement) => inputLines.Where(l => l.StartsWith(movement))
                                        .Select(l => l.ParseInts().Single())
                                        .Sum();

    var horiz = moveAll("forward");
    var vert = moveAll("down") - moveAll("up");

    return horiz * vert;
}
```

```polyglot-notebook
// After following these instructions, you would have a horizontal position of
// 15 and a depth of 10. (Multiplying these together produces 150.)

var testAnswer = DoMovements(testInputLines);
Console.WriteLine(testAnswer);
```

```polyglot-notebook
// Calculate the horizontal position and depth you would have after following
// the planned course. What do you get if you multiply your final horizontal
// position by your final depth?

var part1Answer = DoMovements(inputLines);
Console.WriteLine(part1Answer);
```

```polyglot-notebook
// 1868935 is correct!
Ensure(1868935, part1Answer);
```

Ok, for part 2 it looks like things are a little more _stateful_ than in part 1. Let's just model the positions and aim as variables and compute them instruction-by-instruction.

```polyglot-notebook
int DoMovements2(string[] inputLines)
{
    var horiz = 0;
    var depth = 0;
    var aim = 0;

    Dictionary<string, Action<int>> moveMap = new();
    moveMap["forward"] = (int x) => {
        horiz += x;
        depth += (aim * x);
    };
    moveMap["down"] = (int x) => aim += x;
    moveMap["up"] = (int x) => aim -= x;

    foreach (var line in inputLines)
    {
        var (motion, amount) = line.Split(' ');
        moveMap[motion](int.Parse(amount));
    }

    return horiz * depth;
}
```

```polyglot-notebook
// After following these new instructions, you would have a horizontal position
// of 15 and a depth of 60. (Multiplying these produces 900.)

var part2TestAnswer = DoMovements2(testInputLines);
Console.WriteLine(part2TestAnswer);
```

```polyglot-notebook
// Using this new interpretation of the commands... What do you get if you
// multiply your final horizontal position by your final depth?

var part2Answer = DoMovements2(inputLines);
Console.WriteLine(part2Answer);
```

```polyglot-notebook
// 1965970888 is correct!
Ensure(1965970888, part2Answer);
```
