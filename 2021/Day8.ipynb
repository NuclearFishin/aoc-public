```polyglot-notebook
HashSet<int> easyDigits = [
    // 1 = 2 segments
    2,
    // 4 = 4 segments
    4,
    // 7 = 3 segments
    3,
    // 8 = 7 segments
    7
 ];

 int CountEasyDigits(string[] inputLines)
 {
    return inputLines.Select(countOneLine).Sum();

    int countOneLine(string inputLine)
    {
        var (_, output) = inputLine.Split(" | ").ToArray();
        var digits = output.Split(' ');

        return digits.Where(d => easyDigits.Contains(d.Length)).Count();
    }
 }
```

### --- Part Two ---

You may find the puzzle description at: https://adventofcode.com/2021/day/8

Ok, for this part I think we need set reduction: we know that the easy digits give us good information: they tell us that, for each _actual_ segment, which _possible_ segment it might be. From there, we can

* remove all other possible segments from those actual segments.
* Once we find out a single mapping, we can remove that segment from other possible mappings

Hopefully we can progressively eliminate the possibilities until we find all the mappings.

```polyglot-notebook
//   0:      1:      2:      3:      4:
//  aaaa    ....    aaaa    aaaa    ....
// b    c  .    c  .    c  .    c  b    c
// b    c  .    c  .    c  .    c  b    c
//  ....    ....    dddd    dddd    dddd
// e    f  .    f  e    .  .    f  .    f
// e    f  .    f  e    .  .    f  .    f
//  gggg    ....    gggg    gggg    ....

//   5:      6:      7:      8:      9:
//  aaaa    aaaa    aaaa    aaaa    aaaa
// b    .  b    .  .    c  b    c  b    c
// b    .  b    .  .    c  b    c  b    c
//  dddd    dddd    ....    dddd    dddd
// .    f  e    f  .    f  e    f  .    f
// .    f  e    f  .    f  e    f  .    f
//  gggg    gggg    ....    gggg    gggg

Dictionary<string, char> segmentDigitLookup = new() {
    ["abcefg"] = '0',
    ["cf"] = '1',
    ["acdeg"] = '2',
    ["acdfg"] = '3',
    ["bcdf"] = '4',
    ["abdfg"] = '5',
    ["abdefg"] = '6',
    ["acf"] = '7',
    ["abcdefg"] = '8',
    ["abcdfg"] = '9',
};

const string allSegments = "abcdefg";

var q = from d in segmentDigitLookup.Keys
        group d by d.Length into dGroup
        let dCommon = allSegments.Where(s => dGroup.All(dg => dg.Contains(s)))
        select (length: dGroup.Key, commonDigits: dCommon.ToHashSet());
var commonSegmentsPerDigitLength = q.ToDictionary(qq => qq.length, qq => qq.commonDigits);

commonSegmentsPerDigitLength.Display();
```

```polyglot-notebook
int Unscramble(string line)
{
    var lineDigits = line.ParseAll(@"\w+").ToList();
    var scrambledDigits = lineDigits[..^4];
    var testDigits = lineDigits[^4..];

    var possibleSegments = allSegments.ToDictionary(
        s => s,
        s => allSegments.ToHashSet()
    );

    foreach (var scrambledDigit in scrambledDigits)
    {
        refinePossibleSegments(scrambledDigit);
    }

    eliminate();

    var scrambledToActual = possibleSegments.ToDictionary(
        kv => kv.Value.Single(),
        kv => kv.Key
    );

    string actualDigits = new([..testDigits.Select(unscrambleDigit)]);
    return int.Parse(actualDigits);

    void refinePossibleSegments(string scrambledDigit)
    {
        foreach (var segment in commonSegmentsPerDigitLength[scrambledDigit.Length])
        {
            possibleSegments[segment].IntersectWith(scrambledDigit);
        }
    }

    void eliminate()
    {
        // Progressively eliminate: for any segments we've found, subtract them
        // from the not-yet-found ones
        var foundSegments = possibleSegments.Values.SelectMany(s => s.Count is 1 ? s : []);
        while (foundSegments.Count() < 7)
        {
            foreach (var (actual, possibles) in possibleSegments)
            {
                if (possibles.Count is 1) { continue; }
                possibles.ExceptWith(foundSegments.ToList());
            }
        }
    }

    char unscrambleDigit(string scrambledDigit)
    {
        var actualSegments = scrambledDigit.Select(s => scrambledToActual[s]);
        string actualDigit = new([..actualSegments.OrderBy(s => s)]);
        return segmentDigitLookup[actualDigit];
    }
}
```

```polyglot-notebook
int DecodeAndSum(string[] inputLines) => inputLines.Select(Unscramble).Sum();
```

```polyglot-notebook
// Adding all of the output values in this larger example produces 61229.

var part2TestAnswer = DecodeAndSum(testInputLines);
Console.WriteLine(part2TestAnswer);
```
