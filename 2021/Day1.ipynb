```polyglot-notebook
int CountDecreases(string[] inputLines)
{
    var inputNums = inputLines.Select(int.Parse).ToArray();
    var decreases = inputNums.Zip(inputNums.Skip(1))
        .Count(ab => ab.Second > ab.First);

    return decreases;
}
```

```polyglot-notebook
// In this example, there are 7 measurements that are larger than the previous
// measurement.

var testAnswer = CountDecreases(testInputLines);
Console.WriteLine(testAnswer);
```

```polyglot-notebook
// How many measurements are larger than the previous measurement?

var part1Answer = CountDecreases(inputLines);
Console.WriteLine(part1Answer);
```

Ok, so for part 2 I think we can compute all the 3-digit sliding windows, then use the same comparison stragegy from part 1!

```polyglot-notebook
int CountSlidingDecreases(string[] inputLines)
{
    var numbers = inputLines.Select(int.Parse).ToArray();

    var sum3s = SlidingWindow(numbers, 3, i => i.Sum(), (prev, rem, add) => prev - rem + add).ToList();

    return sum3s.Zip(sum3s[1..]).Count(ab => ab.Second > ab.First);
}
```

```polyglot-notebook
// In this example, there are 5 sums that are larger than the previous sum.

var part2TestAnswer = CountSlidingDecreases(testInputLines);
Console.WriteLine(part2TestAnswer);
```

```polyglot-notebook
// Consider sums of a three-measurement sliding window. How many sums are larger
// than the previous sum?

var part2Answer = CountSlidingDecreases(inputLines);
Console.WriteLine(part2Answer);
```

```polyglot-notebook
// 1653 is correct!
Ensure(1653, part2Answer);
```
