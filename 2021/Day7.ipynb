```polyglot-notebook
// Spoiler alert: this changes in part 2
using FuelCostFunc = System.Func<int, int, int>;

int MinFuel(string[] inputLines, FuelCostFunc fuelCostFunc = null)
{
    fuelCostFunc ??= FuelCost;

    var positions = inputLines[0].ParseInts().ToArray();
    var (min, max) = (positions.Min(), positions.Max());
    var posCount = 1 + max - min;

    var allCosts = Range(min, posCount).Select(h => positions.Sum(n => fuelCostFunc(n, min + h)));
    return allCosts.Min();
}

int FuelCost(int a, int b) => a > b ? a - b : b - a;
```

```polyglot-notebook
// This costs a total of 37 fuel.

var testAnswer = MinFuel(testInputLines);
Console.WriteLine(testAnswer);
```

```polyglot-notebook
// Determine the horizontal position that the crabs can align to using the least
// fuel possible. How much fuel must they spend to align to that position?

var part1Answer = MinFuel(inputLines);
Console.WriteLine(part1Answer);
```

Ok, for part 2 we need a new fuel cost function to calculate the sum of a sequence of integers. There's a well-known mathematical trick for this, so I think this part will be straightforward.

```polyglot-notebook
int SequenceSum(int s) => (s % 2) switch {
    0 => (s + 1) * (s / 2),
    _ => s + SequenceSum(s - 1)
};

int FuelCost2(int a, int b) => SequenceSum(FuelCost(a, b));
```

```polyglot-notebook
// This costs a total of 168 fuel.

var part2TestAnswer = MinFuel(testInputLines, FuelCost2);
Console.WriteLine(part2TestAnswer);
```
